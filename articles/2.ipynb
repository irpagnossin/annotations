Do que precisamos antes de começar...

```python
from scipy.integrate import odeint
from IPython.display import display, Math, Latex
%pylab
%matplotlib inline
```

Começamos declarando a massa do objeto, em unidades arbitrárias. Para facilitar a vida, convém interpretá-la como se fosse dada no Sistema Internacional (SI) de unidades:

```python
m = 1
```

Em seguida, declaramos a energia potencial **total** desse objeto, como função da posição. Neste exemplo, usamos $U(x) = \frac{1}{2}x^4-\frac{2}{3}x^3-2x^2$:

```python
def U(x):
    return 1./2*x**4 - 2./3*x**3 - 2.*x**2
```

Vamos tratar apenas dos problemas para os quais a força que age sobre o corpo é derivada da energia potencial. Ou seja, $F(x) \doteq -dU/dx$. É importante manter a expressão de $U(x)$, acima, e de $F(x)$, abaixo, em conformidade, ou os resultados não serão compatíveis:

```python
def F(x):
    return -2*x**3 + 2*x**2 + 4*x
```

O problema básico de dinâmica que queremos resolver pode ser genericamente representado pela equação diferencial $\ddot x = f(\dot x, x, t)$. Ou seja, a derivada de segunda ordem da posição $x$ é dada pela função $f$, que depende do próprio $x$, de sua derivada de primeira ordem e, eventualmente, do tempo $t$. Por exemplo, no nosso caso estamos interessados em resolver a equação do movimento, que tem origem na segunda lei de Newton:
$$
F(x) = m\ddot x \Rightarrow \ddot x = \frac{F(x)}{m}.
$$

Neste exemplo, $f \equiv F/m$ só depende de $x$.

Mas a biblioteca que temos à disposição só é capaz de resolver problemas do tipo $\dot x = g(x,t)$. Como fazer? Felizmente, qualquer equação diferencial de ordem $N$ pode ser decomposta num sistema de $N$ equações diferenciais de primeira ordem, bastando para isso criar variáveis intermediárias. Por exemplo, na equação acima introduzimos a variável $v \doteq \dot x$ (a velocidade), o que nos permite reescrever a equação diferencial de segunda ordem acima como duas equações diferenciais de primeira ordem, assim:
\begin{align*}
\dot x &= v\\
\dot v &= F/m.
\end{align*}

O que a função <code>oscillator</code> abaixo faz é exatamente definir o lado direito da equação acima: ela recebe como parâmetro de entrada a condição inicial (o passo anterior da iteração) e, para qualquer tempo $t$ posterior, ela devolve o valor de $\dot x$ (primeiro elemento do vetor) e $\dot v$ (segundo), o que permite ao integrador (mais à frente) calcular os valores de $x$ e $v$ no próximo passo.

```python
def oscillator(ic, t):
    x = ic[0]
    v = ic[1]
    return [v, F(x)/m]
```

**Condições iniciais** (ie, em $t = 0$): o primeiro elemento é a posição; o segundo, a velocidade.

```python
ic = [-1.5, 0]
```

Agora, partindo das condições iniciais e das funções que definem o lado direito do sistema de equações diferenciais, a função <code>odeint</code> pode calcular o valor de $x$ e de $v$ em $\Delta t$, $2\Delta t$, etc até o instante de tempo desejado [na configuração abaixo, em <code>linspace</code>, desde $t_0 = 0$ até $t_n = 10$, com uma amostragem de $n = 100$ intervalos $\Delta t = (t_n - t_0)/n = 0.1$].

```python
t0 = 0
tn = 10
n = 100
t = linspace(t0, tn, n+1)
solution = odeint(oscillator, ic, t)
```

O resultado (<code>solution</code>) da operação acima é uma matriz contendo os valores de $x$ na primeira coluna e os de $v$ na segunda. Cada valor é associado aos valores do vetor <code>t</code>, que representa a evolução temporal. Agora podemos desenhar o gráfico das soluções:

```python
x = solution[:,0]
v = solution[:,1]
plot(t, x, label="$x$")
plot(t, v, label="$v$")
legend()
```

Além do gráfico das equações horárias, outro que é muito útil é a trajetória desse sistema no espaço de fases, isto é, no plano $p \times x$, onde $p$ é o momento linear:

```python
p = m*v # Momento linear
plot(x,p)
```

Em tempo, é útil analisar os gráficos acima com o de $U(x)$ à mão (note que a energia mecânica é mostrada em azul):

```python
xmin = -2
xmax = +3
nptos = 100

energy = m*ic[1]**2/2 + U(ic[0]) # Energia mecânica (conservada), calculada com base nas condições iniciais
x = linspace(xmin, xmax, nptos) # Posição
E = repeat(energy, len(x)) # Cria um vetor que contém várias cópias da energia, do mesmo tamanho de x, para poder plotar
plot(x, E, label="Energia")
plot(x, U(x), label="$U(x)$")
legend()
```
